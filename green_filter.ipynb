```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
color_image = cv2.imread("test.jpg")
plt.imshow(color_image)
```

```python
color_image = cv2.cvtColor(color_image,cv2.COLOR_BGR2RGB)
#plt.imshow(color_image)
print(color_image.shape)
```

```python
a = np.copy(color_image)
#plt.imshow(a)
len
```

```python
#Green Mask

a = np.copy(color_image)
lower_green = np.array([60,50,10], dtype=np.uint8)
upper_green = np.array([81,255,20], dtype=np.uint8)

green_mask = cv2.inRange(a,lower_green,upper_green)

#plt.imshow(green_mask,cmap='gray')
a[green_mask == 255] = [255,255,255]
plt.imshow(a)
```

```python
filtered_image = cv2.imwrite("/home/abh300/New_Virtual/new_cv_task/white_filter.jpg",a)
```

```python
masked_image = np.copy(color_image)
masked_image[green_mask == 255] =[0,0,0]
plt.imshow(masked_image)
```

```python
bg_image = cv2.imread("SanFrancisco.jpg")
bg_image = cv2.cvtColor(bg_image,cv2.COLOR_BGR2RGB)
#plt.imshow(bg_image)

crop_bg = bg_image[0:360,0:540]
crop_bg[green_mask == 0] = [0,0,0]
plt.imshow(crop_bg)
```

```python
final_image = crop_bg + masked_image
plt.imshow(final_image)
```
